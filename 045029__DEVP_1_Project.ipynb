The purpose of this project is to automate the extraction and analysis of data of All Formula 1 champions from 1951 to 2021 from the website "wikipedia.com/All Formula 1 champions" The goal is to obtain a comprehensive list of f1 champions, including their teams, Engines,No of poles, and Win %. And do Basic Descriptive & Mathematical or Statistical Analysis.

```python
import requests
from bs4 import BeautifulSoup
from tabulate import tabulate

# Define the URL of the Simple English Wikipedia page
url = "https://simple.wikipedia.org/wiki/List_of_Formula_One_World_Drivers%27_Champions"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the table with the class "wikitable" containing the championship winners
    championship_table = soup.find('table', {'class': 'wikitable'})

    # Initialize an empty list to store the championship data
    championship_data = []

    # Loop through the rows in the table and extract the championship data
    for row in championship_table.find_all('tr'):
        columns = row.find_all(['th', 'td'])
        row_data = [column.text.strip() for column in columns]
        if row_data:  # Ignore empty rows
            championship_data.append(row_data)

    # Use tabulate to format and print the championship data
    headers = championship_data[0]  # First row as headers
    data = championship_data[1:]  # Remaining rows as data

    table = tabulate(data, headers, tablefmt="pipe")
    print(table)
else:
    print("Failed to retrieve the page. Status code:", response.status_code)
```

9. Summary Statistics and Insights for Drivers Age (Mean, median, mode and standard deviation,Variance).

```python
import statistics

# Sample age data
ages = [44, 40, 34, 35, 29, 33, 34, 33, 27, 30, 31, 30, 28, 25, 26, 29, 38, 29, 34, 29, 34, 30, 28, 39, 25, 36, 26, 30, 24, 28, 23, 29, 23, 31, 24]

# Calculate the mean
mean_age = statistics.mean(ages)

# Calculate the median
median_age = statistics.median(ages)

# Calculate the mode (returns the first mode if there are multiple)
try:
    mode_age = statistics.mode(ages)
except statistics.StatisticsError:
    mode_age = "No unique mode found"

# Calculate the interquartile range (IQR)
q1 = statistics.median(ages[:len(ages)//2])
q3 = statistics.median(ages[len(ages)//2:])
iqr = q3 - q1

# Calculate the standard deviation
std_deviation_age = statistics.stdev(ages)

# Calculate the variance
variance_age = statistics.variance(ages)

# Title
title = "Summary Statistics for Age Data"

# Print the title and statistics
print(title)
print("-" * len(title))
print(f"Mean Age: {mean_age:.2f}")
print(f"Median Age: {median_age}")
print(f"Mode Age: {mode_age}")
print(f"IQR: {iqr}")
print(f"Standard Deviation: {std_deviation_age:.2f}")
print(f"Variance: {variance_age:.2f}")
```

12. "Distribution of Ages and Its Components" effectively describes the content of your plot, which displays the distribution of ages and highlights its components, such as the box, whiskers, median line, and outliers. It provides an overview of what the plot represents and what readers can expect to see.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample age data
ages = [44, 40, 34, 35, 29, 33, 34, 33, 27, 30, 31, 30, 28, 25, 26, 29, 38, 29, 34, 29, 34, 30, 28, 39, 25, 36, 26, 30, 24, 28, 23, 29, 23, 31, 24]

# Define custom colors
box_color = 'skyblue'        # Color for the box
whisker_color = 'salmon'     # Color for the whiskers
median_color = 'limegreen'   # Color for the median line
flier_color = 'violet'       # Color for the outliers

# Create a box plot with custom colors
plt.figure(figsize=(8, 6))
bp = plt.boxplot(ages, vert=False, patch_artist=True)

# Customize box plot elements with colors
for element in ['boxes', 'whiskers', 'medians', 'fliers']:
    plt.setp(bp[element], color='black')  # Set edge color for all elements

# Fill the box with a custom color
for patch in bp['boxes']:
    patch.set_facecolor(box_color)

# Set colors for whiskers, median line, and outliers
for line in bp['whiskers']:
    line.set(color=whisker_color, linewidth=2)
for line in bp['medians']:
    line.set(color=median_color, linewidth=2)
for flier in bp['fliers']:
    flier.set(marker='o', color=flier_color, alpha=0.5)

# Calculate box plot statistics
median = np.median(ages)
q1 = np.percentile(ages, 25)
q3 = np.percentile(ages, 75)
iqr = q3 - q1
lower_whisker = q1 - 1.5 * iqr
upper_whisker = q3 + 1.5 * iqr
outliers = [age for age in ages if age < lower_whisker or age > upper_whisker]

# Annotate the plot with statistics
plt.annotate(f'Median: {median}', xy=(median, 1), xytext=(median + 1, 1.2), fontsize=12, color='black')
plt.annotate(f'Q1: {q1}', xy=(q1, 1), xytext=(q1 - 3, 1.2), fontsize=12, color='black')
plt.annotate(f'Q3: {q3}', xy=(q3, 1), xytext=(q3 + 1, 1.2), fontsize=12, color='black')
plt.annotate(f'IQR: {iqr}', xy=(median, 1), xytext=(median - 3, 0.8), fontsize=12, color='black')
plt.annotate(f'Outliers: {outliers}', xy=(median, 1), xytext=(median - 3, 0.6), fontsize=12, color='black')

# Set labels and title
plt.xlabel('Age')
plt.title('Distribution of Ages with Statistics')
plt.grid(True)

# Show the plot
plt.show()
```
