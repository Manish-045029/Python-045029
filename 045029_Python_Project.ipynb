```python
import requests
from bs4 import BeautifulSoup
from tabulate import tabulate

# Define the URL of the Simple English Wikipedia page
url = "https://simple.wikipedia.org/wiki/List_of_Formula_One_World_Drivers%27_Champions"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the table with the class "wikitable" containing the championship winners
    championship_table = soup.find('table', {'class': 'wikitable'})

    # Initialize an empty list to store the championship data
    championship_data = []

    # Loop through the rows in the table and extract the championship data
    for row in championship_table.find_all('tr'):
        columns = row.find_all(['th', 'td'])
        row_data = [column.text.strip() for column in columns]
        if row_data:  # Ignore empty rows
            championship_data.append(row_data)

    # Use tabulate to format and print the championship data
    headers = championship_data[0]  # First row as headers
    data = championship_data[1:]  # Remaining rows as data

    table = tabulate(data, headers, tablefmt="pipe")
    print(table)
else:
    print("Failed to retrieve the page. Status code:", response.status_code)
```

```python
from tabulate import tabulate
from colorama import Fore, Style

# Define the data
data = [
    ["United Kingdom United Kingdom", 10, 20],
    ["Germany Germany", 3, 12],
    ["Brazil Brazil", 3, 8],
    ["Argentina Argentina", 1, 5],
    ["France France", 1, 4],
    ["Australia Australia", 2, 4],
    ["Austria Austria", 2, 4],
    ["Finland Finland", 3, 4],
    ["Italy Italy", 2, 3],
    ["United States United States", 2, 2],
    ["Netherlands Netherlands", 1, 2],
    ["Spain Spain", 1, 2],
    ["South Africa South Africa", 1, 1],
    ["Canada Canada", 1, 1],
    ["New Zealand New Zealand", 1, 1]
]

# Define headers
headers = ["Country", "Drivers", "Total wins"]

# Function to color rows based on conditions
def color_rows(row):
    if row[1] >= 3:
        return [Fore.GREEN + str(cell) + Style.RESET_ALL if isinstance(cell, int) else cell for cell in row]
    elif row[2] >= 10:
        return [Fore.RED + str(cell) + Style.RESET_ALL if isinstance(cell, int) else cell for cell in row]
    else:
        return row

# Apply the color function to each row
colored_data = [color_rows(row) for row in data]

# Display the data in tabulate format
table = tabulate(colored_data, headers, tablefmt="grid")

# Print the table
print(table)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample dataset of engines and their total wins
engines_data = [
    {"Engine": "Ferrari", "Total Wins": 15},
    {"Engine": "McLaren", "Total Wins": 12},
    {"Engine": "Mercedes GP", "Total Wins": 8},
    {"Engine": "Williams", "Total Wins": 7},
    {"Engine": "Red Bull", "Total Wins": 6},
    {"Engine": "Lotus", "Total Wins": 6},
    {"Engine": "Brabham", "Total Wins": 4},
    {"Engine": "Cooper", "Total Wins": 2},
    {"Engine": "Renault", "Total Wins": 1},
]

# Filter out engines with zero wins
engines_data = [engine for engine in engines_data if engine["Total Wins"] > 0]

# Extract engine names and total wins
engine_names = [engine["Engine"] for engine in engines_data]
total_wins = [engine["Total Wins"] for engine in engines_data]

# Set the number of engines
num_engines = len(engine_names)

# Create an array of indices for the engines
indices = np.arange(num_engines)

# Define a list of colors for the bars
colors = ['red', 'orange', 'silver', 'lightskyblue', 'darkblue', 'maroon', 'lightseagreen', 'pink', 'yellow']

# Create a bar chart with different colors for the bars
plt.figure(figsize=(12, 6))
bars = plt.bar(indices, total_wins, color=colors)

# Set the labels for the x-axis and y-axis
plt.xlabel('Engines')
plt.ylabel('Total Wins')

# Set the title for the bar chart
plt.title('Total Wins by Engine (Engines with Wins > 0)')

# Set the x-axis tick labels and their positions
plt.xticks(indices, engine_names, rotation=45, ha='right')

# Add labels to the bars
for bar, win_count in zip(bars, total_wins):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() + 0.2, str(win_count), fontsize=10, fontweight='bold')

# Display the bar chart
plt.tight_layout()
plt.show()
```

```python
import statistics

# Sample age data
ages = [44, 40, 34, 35, 29, 33, 34, 33, 27, 30, 31, 30, 28, 25, 26, 29, 38, 29, 34, 29, 34, 30, 28, 39, 25, 36, 26, 30, 24, 28, 23, 29, 23, 31, 24]

# Calculate the mean
mean_age = statistics.mean(ages)
print(f"Mean Age: {mean_age:.2f}")

# Calculate the median
median_age = statistics.median(ages)
print(f"Median Age: {median_age}")

# Calculate the mode
mode_age = statistics.mode(ages)
print(f"Mode Age: {mode_age}")

# Calculate the standard deviation
std_deviation_age = statistics.stdev(ages)
print(f"Standard Deviation: {std_deviation_age:.2f}")

# Calculate the variance
variance_age = statistics.variance(ages)
print(f"Variance: {variance_age:.2f}")

# Calculate the interquartile range (IQR)
q1 = statistics.median(ages[:len(ages)//2])
q3 = statistics.median(ages[len(ages)//2:])
iqr = q3 - q1
print(f"IQR: {iqr}")
```
